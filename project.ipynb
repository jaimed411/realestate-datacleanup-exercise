#### Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (★★★) 

```python
mean_price = data_frame['price'].mean()
mean_rooms = data_frame['rooms'].mean()
mean_surface = data_frame['surface'].mean()
mean_bathrooms = data_frame['bathrooms'].mean()

var_price = data_frame['price'].var()
var_rooms = data_frame['rooms'].var()
var_surface = data_frame['surface'].var()
var_bathrooms = data_frame['bathrooms'].var()

print("Media Muestral:")
print("-Precio:", mean_price, "€")
print("-Habitaciones:", mean_rooms, "habitaciones")
print("-Superficie:", mean_surface, "m2")
print("-Baños:", mean_bathrooms)

print("\nVarianza:")
print("-Precio:", var_price, "€")
print("-Habitaciones:", var_rooms, "habitaciones")
print("-Superficie:", var_surface, "m2")
print("-Baños:", var_bathrooms)
```

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
filtrar_data = data_frame[data_frame['level5'].isin(poblaciones)]

for population in poblaciones:
    population_data = filtrar_data[filtrar_data['level5'] == population]
    precio_normalizado = (population_data['price'] - population_data['price'].mean()) / population_data['price'].std()
    plt.hist(precio_normalizado, bins=40, label=population, histtype = "step")

plt.title('Precios normalizados de Fuelabrada, Leganés, Getafe y Alcorcón',
          fontdict={'fontsize':10, 'color': '#262526'})
plt.xlabel('Precio normalizado')
plt.ylabel('Frequencia')
plt.legend()
plt.show()

```

```python
'''# Run in console the following command before run the rest of code:
# jupyter nbextension enable --py --sys-prefix ipyleaflet

from ipyleaflet import Map, basemaps

# Map centred on (60 degrees latitude et -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (40.265091669716746, -3.6731693949274766), zoom = 10, min_zoom = 1, max_zoom = 20, basemap=basemaps.Stamen.Terrain)

map
'''

'''## con geopandas y contextily. tiene q tener un archivo .geojson
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

casas = "assets/map (2).geojson"    
map_data = gpd.read_file(casas)

fig, ax = plt.subplots(figsize=(10, 10))
 
ax.set_title('Viviendas zona sur Madrid',  
             fontdict={'color': '#b81425'})
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')

map_data.plot(cmap='winter', ax=ax)

ctx.add_basemap(ax, crs=map_data.crs.to_string(), source=ctx.providers.Stamen.Terrain)

plt.show()'''

from ipyleaflet import Map, MagnifyingGlass, basemaps, CircleMarker

# Create a map object
map = Map(center=(40.285113275482985, -3.6120606264778465), zoom=11, min_zoom=1, max_zoom=20, basemap=basemaps.Stamen.Terrain)

# List of values to subset
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón", "Pinto", "San Martín de la Vega", "Arganda del Rey"]

# Define a dictionary to map values to marker colors
color_poblaciones_sur = {"Fuenlabrada": "cadetblue", "Leganés": "green", "Getafe": "red", "Alcorcón": "darkorange",
               "Pinto": "blue", "San Martín de la Vega": "purple", "Arganda del Rey": "yellow"}

# Create a boolean mask for filtering
mascara_level5 = data_frame['level5'].isin(poblaciones)

# Apply the mask to the DataFrame to create the subset
subset_df = data_frame[mascara_level5]

# Filter out rows with 0 latitude and 0 longitude
subset_df = subset_df[(subset_df['latitude'] != '0') & (subset_df['longitude'] != '0')]

# Use list comprehension to create tuples from latitude and longitude columns
latitud_list = [float(lat.replace(',', '.')) for lat in subset_df['latitude']]
longitud_list = [float(long.replace(',', '.')) for long in subset_df['longitude']]

# Loop through the latitude and longitude lists and add circle markers to the map
for i, (lat, long) in enumerate(zip(latitud_list, longitud_list)):
    population = subset_df.iloc[i]['level5']
    color = color_poblaciones_sur.get(population, "black")
    marker = CircleMarker(location=(lat, long), radius=1, color=color, fill_color=color)
    map.add_layer(marker)

# Add Magnifying Glass
magnifying_glass = MagnifyingGlass(radius=40, zoom_offset=4)
map.add_control(magnifying_glass)

# Display the map
map

```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
ciudades = ['Getafe', 'Alcorcón']
filtrar_data = data_frame[data_frame['level5'].isin(ciudades)]

getafe_data = filtrar_data[filtrar_data['level5'] == 'Getafe']
getafe_preciomax_m2 = getafe_data['price'].max()
getafe_preciomin_m2 = getafe_data['price'].min()

alcorcon_data = filtrar_data[filtrar_data['level5'] == 'Alcorcón']
alcorcon_preciomax_m2 = alcorcon_data['price'].max()
alcorcon_preciomin_m2 = alcorcon_data['price'].min()

print("Precio maximo m2 en Getafe:", getafe_preciomax_m2, "€")
print("Precio minimo m2 en Getafe:", getafe_preciomin_m2, "€")

print("\nPrecio maximo m2 en Alcorcón:", alcorcon_preciomax_m2, "€")
print("Precio minimo m2 en Alcorcón:", alcorcon_preciomin_m2, "€")
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

```python
selected_towns = ['Valdemorillo', 'Galapagar']
filtered_data = data_frame[data_frame['level5'].isin(selected_towns)]

valdemorillo_data = filtered_data[filtered_data['level5'] == 'Valdemorillo']
valdemorillo_precio_m2 = valdemorillo_data['price'] / valdemorillo_data['surface']
valdemorillo_precio_promedio_m2 = valdemorillo_precio_m2.mean()

galapagar_data = filtered_data[filtered_data['level5'] == 'Galapagar']
galapagar_precio_m2 = galapagar_data['price'] / galapagar_data['surface']
galapagar_precio_promedio_m2 = galapagar_precio_m2.mean()

diferencia_precio = valdemorillo_precio_promedio_m2 - galapagar_precio_promedio_m2

print("Precio promedio por m2 en Valdemorillo:", valdemorillo_precio_promedio_m2, "€")
print("Precio promedio por m2 en Galapagar:", galapagar_precio_promedio_m2, "€")
print("Hay una diferencia de precio promedio de", diferencia_precio, "€, por lo que los precios no son los mismos.")
```

#### Make the same plot for 4 different populations (level5) and rearrange them on the same graph? (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
ciudades = ["Las Rozas de Madrid", "Navalcarnero", "Parla", "Alcalá de Henares"]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('SUPERFICIE VS PRECIO')

for i, population in enumerate(ciudades):
    ciudades_data = data_frame[data_frame['level5'] == population]
    surperficie_area = ciudades_data['surface']
    precio = ciudades_data['price']
    
    row = i // 2
    col = i % 2
    
    axs[row, col].scatter(surperficie_area, precio)
    axs[row, col].set_title(population)
    #axs[row, col].set_xlabel('Superficie')
    #axs[row, col].set_ylabel('Precio')

plt.show()


fig, ax = plt.subplots(figsize=(5, 5))
fig.suptitle('SUPERFICIE VS PRECIO DE 4 CIUDADES')

for population in ciudades:
    ciudades_data = data_frame[data_frame['level5'] == population]
    surperficie_area = ciudades_data['surface']
    precio = ciudades_data['price']
    
    ax.scatter(surperficie_area, precio, label=population, alpha=0.2)

ax.set_xlabel('Superficie')
ax.set_ylabel('Precio')
ax.legend()
plt.show()


```
